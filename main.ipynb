Data loading

```python
import os
import random
from scipy.io import loadmat

base_path = r'F:\code\Final-project_radar_navigation\Dataset'  # 替换成你的总文件夹路径
all_data = []

for folder_name in os.listdir(base_path):
    sub_folder_path = os.path.join(base_path, folder_name)
    if os.path.isdir(sub_folder_path):
        mmwave_path = os.path.join(sub_folder_path, 'mmwave')
        if os.path.exists(mmwave_path) and os.path.isdir(mmwave_path):
            for file_name in os.listdir(mmwave_path):
                if file_name.endswith('.mat'):
                    mat_file_path = os.path.join(mmwave_path, file_name)
                    data = loadmat(mat_file_path)
                    all_data.append((folder_name, file_name, data))

num_samples = 3  # 要随机查看的样本数
samples = random.sample(all_data, min(len(all_data), num_samples))

for folder, file, data in samples:
    print(f"Folder: {folder}, File: {file}")
    for key in ['I', 'Q', 't', 'RawPoints', 'rawImage_XYZ']:  # 只关注特定的变量
        if key in data:
            print(f"Variable name: {key}")
            print(f"Data sample (first 5 elements): {data[key][:5] if data[key].size > 0 else 'Empty'}")

```
